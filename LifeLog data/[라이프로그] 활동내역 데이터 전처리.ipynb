* Target data 는 DIAG_NM 컬럼으로 CN, MCI, DEM 존재 (무증상치매, 경도인지장애, 치매)

```python
df['DIAG_NM'].value_counts()
```

* Dementia 데이터는 571개로 약 14명의 사용자만 치매를 가지고 있음 

```python
import numpy as np

email = np.array(df['EMAIL'])
target = np.array(df['DIAG_NM'])

target_dict = dict(np.c_[email,target]) # Grouping 이후 라벨링 하기 위한 딕셔너리 생성
```

```python
drop_cols = df.describe(include = 'O').columns[1:] # 불필요한 행을 제거
                                                    # 해당 object 컬럼들은 모두 수치형 변수로 표현 되어 있음

print(drop_cols)

df = df.drop(drop_cols,axis = 1) 
```

```python
df = df.groupby('EMAIL').mean().reset_index() # 그룹핑
```

```python
df['target'] = df['EMAIL'].map(target_dict)
```

```python
df = df.drop('EMAIL',axis = 1) # index 역할하던 email 제거 
```
